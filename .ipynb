```python
from itertools import product

# Tìm tất cả các tổ hợp 6 chữ số có tổng chia 10 dư 4
valid_numbers = []
for digits in product(range(10), repeat=6):
    if sum(digits) % 10 == 4:
        valid_numbers.append("".join(map(str, digits)))

# In ra 10 kết quả đầu tiên
print(valid_numbers)
```
