```python
from itertools import product

# Tm tt c cc t hp 6 ch s c tng chia 10 d 4
valid_numbers = []
for digits in product(range(10), repeat=6):
    if sum(digits) % 10 == 4:
        valid_numbers.append("".join(map(str, digits)))

# In ra 10 kt qu u tin
print(valid_numbers)
```
